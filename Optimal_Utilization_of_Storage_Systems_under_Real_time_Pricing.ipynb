```python
## Langrangian Cost Function
def L(X,lam,P):
    (1/P.size)*np.dot(P-lam,X) + (1/(alpha*P.size))*np.dot(X,X)
```

```python
X_global1 = []
```

```python
sum_price1 = []
```

```python
for i in range(5):  
    from numpy.random import rand
    # seed random number generator
#     seed(1)
    # generate random numbers between 0-1
    lam_in = rand(1)
#     print(lam_in.shape)

    # price_list.size

    laam = lam_in[0]
    print(laam)

    # laam

    X_incremental = []

    L_incremental = []

#     price_list

    for i in range(744):
        x = alpha*(laam - price_list[i])
        if x <= -B_out:
            x = -B_out
        elif x >= B_in:
            x = B_in
        X_incremental.append(x)
        laam = laam - 6*x
        L_incremental.append(laam)

#     X_incremental
    X_global1.append(X_incremental)

#     len(X_incremental)

    price_incremental = [price_list[i]*X_incremental[i] for i in range(nll.size)]
    sum_price1.append(sum(price_incremental))
```
